```python
class LinearRegCustom:
    
    def __init__(self, lr=1, max_steps=10):
        self.lr = lr
        self.max_steps = max_steps
        
    def fit(self, X, y):
        self.coef_ = np.random.random()
        
        for i in range(self.max_steps):
            self.gradient_decend(X, y)
        
    def gradient_decend(self, X, y):
        delta_coef = self.gradient(X, y)
#         print(delta_coef)
        self.coef_ += delta_coef
        
    def gradient(self, X, y):
        yp = self.predict(X)
        d_e2m = np.mean((yp -y) * X)
        
        return -d_e2m * self.lr
        
    def predict(self, X):
        return X * self.coef_
```

```python
model = LinearRegCustom(lr=.00001, max_steps=20000)

model.fit(X,y)

model.coef_

plt.scatter(X,y)

plt.scatter(X,model.predict(X) )
```
